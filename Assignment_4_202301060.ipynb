```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the provided CSV URL
data_url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(data_url)

# Extract the month-wise total profit data
months = data['month_number']
total_profit = data['total_profit']

# Plotting the line plot
plt.plot(months, total_profit, marker='o', linestyle='-')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data_url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(data_url)
data_without_months = data.drop('month_number', axis=1)
attribute_totals = data_without_months.sum()
attribute_totals.plot(kind='bar')
plt.title('Total Sales and Profit by Feature')
plt.xlabel('Features')
plt.ylabel('Amount')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the provided CSV URL
data_url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(data_url)

# Extracting the first four attributes
attributes = data.columns[1:5]

# Plotting the scatter plot for each attribute with different colors
colors = ['black', 'green', 'pink', 'orange']
for attribute, color in zip(attributes, colors):
    plt.scatter(data['month_number'], data[attribute], color=color, label=attribute)

plt.title('Scatter Plot of First 4 Attributes')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
