```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
data_url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(data_url)

# Plotting all product sales data using Seaborn
plt.figure(figsize=(10, 6))
for product in data.columns[1:7]:  # Selecting product sales columns
    sns.lineplot(x='month_number', y=product, data=data, label=product)

plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Products')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.drop('month_number', axis=1).sum().plot(kind='bar')
plt.title('Total Sales and Profit by Feature (Seaborn)')
plt.ylabel('Amount')
plt.xlabel('Features')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

# Scatter plot using Plotly
fig = px.scatter(data, x='month_number', y=data.columns[1], color=data.columns[1])
fig.update_layout(title='Scatter Plot of First Attribute (Plotly)')
fig.show()


```

```python
# Bar chart using Plotly
fig = px.bar(data.drop('month_number', axis=1).sum(), x=data.drop('month_number', axis=1).columns, y=data.drop('month_number', axis=1).sum().values)
fig.update_layout(title='Total Sales and Profit by Feature (Plotly)', xaxis_title='Features', yaxis_title='Amount')
fig.show()
```

```python
# Histogram using Plotly
fig = px.histogram(data, x='total_profit', nbins=30)
fig.update_layout(title='Histogram of Total Profit (Plotly)', xaxis_title='Total Profit', yaxis_title='Frequency')
fig.show()
```
