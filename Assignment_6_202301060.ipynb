```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
from google.colab import files
uploaded= files.upload()
df = files.upload()
df= pd.read_csv('DATA.csv')
a=df['X']
b=df['Y']
data = list(zip(a, b))
print(data)
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
plt.scatter(a, b, c=kmeans.labels_)
plt.show()
```
